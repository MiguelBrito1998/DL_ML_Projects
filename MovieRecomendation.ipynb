```python
import numpy as np 
import pandas as pd 
import difflib  #libreria especializada en comparar secuencias (archivos de texto, lineas o listas)
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#cargando los datos del dataFRame
movies_data=pd.read_csv(r'C:\Users\USUARIO\Desktop\PythonSiddhardhan\Conjuntosdedatos\movies.csv')
```

```python
#imprimiendo las primeras 5 columnas del dataframe
movies_data['genres']
```

```python
movies_data['genres'].value_counts()
```

```python
#numero de filas y columnas
movies_data.isna().sum()
```

```python
#Selecting the relevant features for recommendation
selected_features=['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#replacing the null values with null strinh

for feature in selected_features:
    movies_data[feature]=movies_data[feature].fillna('')
```

```python
movies_data.isna().sum()
```

```python
combined_features=movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
combined_features.head()
```

```python
#converting the text data to feature vectors
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
#OBTENIENDO LA METRICA COSINE SIMILARITY
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
similarity[2]
```

```python
#Getting name from the user

movie_name=input(' Enter your favorite movie name: ')
print(movie_name)
```

```python
#creando una lista con todas las peliculas relacionadas a la entrada
list_of_all_tittles=movies_data['title'].tolist()
print(list_of_all_tittles)
```

```python
#finding the close mathc for the movie name given by the user

find_close_match=difflib.get_close_matches(movie_name,list_of_all_tittles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
#Find the index of the movie with tittle

index_of_the_movie=movies_data[movies_data.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#Obteniendo lista de peliculas similares
similarity_score=list(enumerate(similarity[index_of_the_movie]))
```

```python
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#Ordenando las peliculas basado en el coseno de similitud

sorted_similar_movies = sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#print the name of similar movies based on the index

print(' Movies suggested for you: \n')

i=1
for movie in sorted_similar_movies:
    index=movie[0]
    similaridad=movie[1]
    title_from_index=movies_data[movies_data.index==index]['title'].values[0]
    if (i<10):
        print(i,".",title_from_index, ", con una similaridad de :" ,similaridad )
        i+=1
```
