```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
import sklearn.datasets 
from xgboost import XGBRegressor
from sklearn import metrics
from sklearn.metrics import accuracy_score 
from sklearn.model_selection import train_test_split
```

Importando el dataset de boston houses

```python
house_price_data=sklearn.datasets.fetch_california_housing()
```

```python
house_price_data
```

```python
#Loading the dataset to a pandas dataFRam
house_price_dataset=pd.DataFrame(house_price_data.data,columns=house_price_data.feature_names)
```

```python
house_price_dataset
```

```python
house_price_dataset['Price']=house_price_data.target
```

```python
house_price_dataset
```

```python
#chqueando de nuevo las columnas y filas de la tabla

house_price_dataset.head()
```

```python
#Determinando si hay valores vacios
house_price_dataset.isnull().sum()
```

```python
#Caracateristicas estadisticas del dataset
house_price_dataset.describe()
```

```python
#correlcion de los datos, aqui se revisara la relacion que hay entre las variables y el target

correlation=house_price_dataset.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, annot=True, cbar=True, fmt='.1f', square=True, annot_kws={'size':8}, cmap='Blues')
```

```python
#Separando los datos
X=house_price_dataset.drop(['Price'],axis=1)
Y=house_price_dataset['Price']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

Model Training
XGBoost Regressor

```python
#cargando el modelo
model=XGBRegressor()
```

```python
#entrenando el modelo
model.fit(X_train,Y_train)
```

```python
#Evaluacion del modelo
#Accuracy score for prediction on training data
training_data_prediction=model.predict(X_train)

```

```python
# accuracy_score(training_data_prediction,Y_train)
Y_train
```

```python
#R squared error
score_1=metrics.r2_score(Y_train, training_data_prediction)

score_2=metrics.mean_absolute_error(Y_train,training_data_prediction)
```

```python
print("R squared error:", score_1 ,"- Mean absolute error:", score_2)
```

```python
#Visualizando los precios actuales y predecidos
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Precios actuales")
plt.ylabel("Predicciones")
plt.title("Precios actuales vs precios predecidos")
plt.show()
```
